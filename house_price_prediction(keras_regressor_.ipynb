```
# df = df[df.line_race != 0]
data2 = data[data.price<1200000]
len(data2)
# data2.head().T
```

```
# droping the unnecessary columns
data2 = data2.drop('id', axis=1)
data2 = data2.drop('zipcode', axis=1)
# data2.info()
```

```
# getting the year and month from the data
data2['date'] = pd.to_datetime(data2['date'])
data2['month'] = data2['date'].apply(lambda date: date.month)
data2['year'] = data2['date'].apply(lambda date: date.year)
# data2.info()
```

```
data2 = data2.drop('date', axis=1) # then dropping the data column
# data2.info()
```

```
data2.groupby('month').mean()
```

```
# data visualization house price vs months and years
fig = plt.figure(figsize=(16,5))
fig.add_subplot(1,2,1)
data2.groupby('month').mean()['price'].plot()
fig.add_subplot(1,2,2)
data2.groupby('year').mean()['price'].plot()
plt.show()
```

```
from sklearn.model_selection import train_test_split

x = data2.drop('price', axis=1).values
y = data2['price'].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```
from sklearn.preprocessing import StandardScaler
s_scaler  = StandardScaler()
x_train = s_scaler.fit_transform(x_train.astype(np.float)) # fit and transform the data
x_test = s_scaler.transform(x_test.astype(np.float)) # only transform the data
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
```

```
model = Sequential()
model.add(Dense(19, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```
model.fit(x_train, y_train, validation_data=(x_test, y_test), batch_size=128, epochs=150)
```

```
loss_df = pd.DataFrame(model.history.history)
loss_df.plot(figsize=(13, 8))
```

```
y_prediction = model.predict(x_test)
```

```
# Our predictions
fig = plt.figure(figsize=(10,5))
plt.scatter(y_test,y_prediction)
# Perfect predictions
plt.plot(y_test,y_test,'r')
```

```
plt.figure(figsize=(13, 8))
res = abs(y_test - y_prediction)
sns.distplot(res, kde=False)
```

```
#single prediction for realtime use
single_pred = data2.drop('price', axis = 1).iloc[1]
single_pred = s_scaler.transform(single_pred.values.reshape(-1, 19))
pred = model.predict(single_pred)
```
